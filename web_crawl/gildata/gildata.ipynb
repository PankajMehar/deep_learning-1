```python
import os
import pandas as pd
import numpy as np
from os.path import join, getsize
```

```python
path = '/home/weiwu/share/deep_learning/data/gildata/'
target_columns = ['table_name_zh', 'table_name_en', '序号', '列名', '中文名称',
         '类型', '空否', '备注', '说明解释', '业务唯一性', '表数据更新频率']
for (root, dirs,
     files) in os.walk(path):
    for filename in files:
        file_path = join(root, filename)
        if file_path.endswith('.xlsx'):
            print(file_path)
            fout = file_path[:-4] + 'csv'
            df = pd.read_excel(file_path)
            df.rename({'talbe_name_zh':'table_name_zh'}, axis=1, inplace=True)
            all_columns = df.columns
            df['备注'].replace({'从表':np.nan}, inplace=True)
            df_notes = df[df['备注'].notnull()]['备注']
            dict_notes_loc = dict(zip(df_notes.unique(), [all_columns.get_loc(x) for x in df_notes.unique()]))
            df_notes.replace(dict_notes_loc, inplace=True)
            for item in df_notes.iteritems():
                # print(df.iloc[item])
                df['备注'].iloc[item[0]] = df.iloc[item]
                df[target_columns].sort_values(['table_name_zh', '序号']).to_csv(fout, encoding='utf_8_sig')
```
